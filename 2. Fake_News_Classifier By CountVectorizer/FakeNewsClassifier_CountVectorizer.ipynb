```python
#Preprocessing
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
WordNet = WordNetLemmatizer()
corpus = []
```

```python
import re
for i in range(len(messages)):
    clean = re.sub('[^a-zA-Z]', ' ', str(messages['title'][i]))
    clean = clean.lower()
    clean = clean.split()
    
    clean = [WordNet.lemmatize(word) for word in clean if word not in stopwords.words('english')]
    clean = ' '.join(clean)
    corpus.append(clean)
```

```python
corpus
```

```python
#Apply bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500,ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
y = messages['label']
```

```python
y
```

```python
#Divide dataset for test and train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.30, random_state = 0 )
```

```python
X_train
```

### Training Model Using Naive Bayes Classifier

```python

from sklearn.naive_bayes import MultinomialNB
FakeNews = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = FakeNews.predict(X_test)
```

```python
y_pred
```

```python
#Comparing Prediction and Test Data -- Confusion Matrix

from sklearn.metrics import confusion_matrix
Confustion_M = confusion_matrix(y_test, y_pred)
```

```python
# Check the Accuracy
from sklearn.metrics import accuracy_score
Accuracy = accuracy_score(y_test, y_pred)
```
